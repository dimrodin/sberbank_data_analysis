### Упражнение 5
* Напишите функцию, которая берет многомерный вещественный массив с пропущенными значениями (np.nan) и возвращает его копию с заполненными пропусками (заменять средним значением всех элементов массива). Если в массиве одни пропущенные значения, заполните их нулями.

````python
def not_nan(arr):
    if np.nansum(arr) == 0:
        arr[np.isnan(arr)] = 0
    else:
        arr[np.isnan(arr)] = np.nanmean(arr)
    return arr
````

````python
b = np.arange(1, 5, 2).reshape(2, )
b
````

````python
a = np.random.random(size=(4, 4))
b = np.random.randint(0, 4, size=(np.random.randint(0, 4)))
a[b] = np.nan
not_nan(a)
````

### Упражнение 12
* Напишите игру [life](https://en.wikipedia.org/wiki/Conway%27s_Game_of_Life), используя массивы numpy.
* Визуализируйте интерактивно ее результаты.



````python
from IPython import display
import time
import matplotlib.pyplot as plt
%matplotlib inline 
````

````python
def life_numpy(X, life_period):
    n, m = X.shape
    palette = np.array([[255, 255, 255], #белый
                        [0, 255, 0]])     #красный
    plt.figure(figsize=X.shape)
    plt.imshow(palette[X])
    plt.grid(True)
    plt.show()
    Y = X.copy()
    while life_period > 0:
        for i in range(n):
            for j in range(m):
                neigh = np.matrix(X[max(0, i-1):min(n, i+2), max(0, j-1):min(m, j+2)])
                if ((np.sum(neigh) - X[i, j]) < 2 or (np.sum(neigh) - X[i, j]) >= 4) and X[i,j] == 1:
                    Y[i, j] = 0
                elif (np.sum(neigh) - X[i, j]) == 3:
                    Y[i,j] = 1
        
        X = Y.copy()
        display.clear_output()
        plt.figure(figsize=X.shape)
        plt.imshow(palette[X])
        plt.grid(True)
        plt.show()
        time.sleep(1)
#        print(Y)
        life_period -= 1
        
````

````python
X = np.zeros(64, dtype=int).reshape(8, 8)
X[2:5, 3:6] = 1
life_period = 10
palette = np.array([[255, 255, 255], #белый
                        [0, 255, 0]])     #зеленый
life_numpy(X, life_period)
````
